## Step 3: ML Functions (Filtering + Scoring + Classification)

```python
def recommend_games(ram, gpu, cpu, genre='all', top_n=10):
    """PC specs ke basis pe games recommend karta hai"""

    # ML Concept 1: Rule-Based Filtering
    # Sirf wahi games rakhno jo user ke PC pe chal sakein
    filtered = []
    for game in GAMES:
        genre_ok = (genre == 'all') or (game['genre'] == genre)
        can_run  = (ram >= game['min_ram']) and (gpu >= game['min_gpu']) and (cpu >= game['min_cpu'])
        if genre_ok and can_run:
            filtered.append(game)

    # ML Concept 2: Weighted Scoring Algorithm
    # Har game ko score do — kitna acha chalega
    scored = []
    for game in filtered:
        gpu_margin = gpu - game['min_gpu']                      # GPU extra power
        ram_margin = (ram - game['min_ram']) / game['min_ram']  # RAM extra %
        cpu_margin = cpu - game['min_cpu']                      # CPU extra power
        score = (gpu_margin * 3) + (ram_margin * 2) + (cpu_margin * 1)
        scored.append({'game': game, 'score': score})

    # ML Concept 3: Ranking — Best match pehle
    scored.sort(key=lambda x: x['score'], reverse=True)

    return scored[:top_n]


def get_compatibility(score):
    """ML Concept 4: Threshold Classification"""
    if score >= 6:   return '✅ Runs Great'
    elif score >= 2: return '🟡 Runs OK'
    else:            return '⚠️  Barely Runs'


def get_settings(score):
    """Score se recommended graphics settings"""
    if score >= 5:   return 'Ultra / Max'
    elif score >= 3: return 'High'
    elif score >= 1: return 'Medium'
    else:            return 'Low'


print('ML functions ready!')
```

## Step 4: Interactive Widget — Dropdown se Select Karo

```python
# Dropdowns banao
ram_dd = widgets.Dropdown(
    options=[('2 GB',2),('4 GB',4),('8 GB',8),('16 GB',16),('32 GB',32)],
    value=8, description='RAM:', style={'description_width':'100px'},
    layout=widgets.Layout(width='340px')
)
gpu_dd = widgets.Dropdown(
    options=[('No Dedicated GPU',0),('GT 710/730',1),('GTX 750 Ti / RX 550',2),
             ('GTX 1050 Ti / RX 570',3),('GTX 1060 / RX 580',4),
             ('GTX 1070 / RX 5700',5),('RTX 2070 / RX 6700 XT',6),
             ('RTX 3080 / RX 6800 XT',7),('RTX 4090 / RX 7900 XTX',8)],
    value=4, description='GPU:', style={'description_width':'100px'},
    layout=widgets.Layout(width='340px')
)
cpu_dd = widgets.Dropdown(
    options=[('Pentium/Celeron',1),('Core i3 / Ryzen 3',2),('Core i5 / Ryzen 5',3),
             ('Core i7 / Ryzen 7',4),('Core i9 / Ryzen 9',5)],
    value=3, description='CPU:', style={'description_width':'100px'},
    layout=widgets.Layout(width='340px')
)
genre_dd = widgets.Dropdown(
    options=[('All Genres','all'),('Action/Shooter','action'),('RPG/Open World','rpg'),
             ('Strategy/Sim','strategy'),('Horror/Survival','horror'),
             ('Sports/Racing','sports'),('Indie/Casual','indie')],
    value='all', description='Genre:', style={'description_width':'100px'},
    layout=widgets.Layout(width='340px')
)
num_dd = widgets.IntSlider(
    value=8, min=3, max=20, description='Results:',
    style={'description_width':'100px'}, layout=widgets.Layout(width='340px')
)

btn = widgets.Button(description='🔍 Find Games For My PC',
    button_style='success', layout=widgets.Layout(width='250px', height='42px'))
out = widgets.Output()

def on_click(b):
    with out:
        out.clear_output()
        results = recommend_games(ram_dd.value, gpu_dd.value, cpu_dd.value, genre_dd.value, num_dd.value)
        print(f'\n  PC: RAM={ram_dd.value}GB | GPU={GPU_NAMES[gpu_dd.value]} | CPU={CPU_NAMES[cpu_dd.value]}')
        print('  ' + '─'*60)
        if not results:
            print('  Koi game nahi mila. Genre All karo ya specs check karo.')
        else:
            print(f'  {len(results)} Games Jo Aapke PC Pe Chalenge:\n')
            for i, item in enumerate(results, 1):
                g = item['game']
                s = item['score']
                bar = '█' * min(int(s), 15) + '░' * (15 - min(int(s), 15))
                print(f"  {i:2}. {g['emoji']}  {g['title']}")
                print(f"       {g['desc']}")
                print(f"       {get_compatibility(s)}  |  {get_settings(s)} Settings")
                print(f"       [{bar}] {s:.1f}\n")

btn.on_click(on_click)

display(HTML('<h2 style="color:#7c3aed">🎮 PC Game Recommender</h2>'),
        HTML('<p style="color:gray">Specs select karo aur button dabao</p>'),
        ram_dd, gpu_dd, cpu_dd, genre_dd, num_dd, btn, out)
print('Widget ready! Upar dropdowns mein apni specs select karo.')
```
